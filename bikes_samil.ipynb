```python
# Convert 'Start Time' column to datetime format
df1['Start Time'] = pd.to_datetime(df1['Start Time'])
df2['Start Time'] = pd.to_datetime(df2['Start Time'])
df3['Start Time'] = pd.to_datetime(df3['Start Time'])
```

```python
# Create DataFrames by month
df_month1 = df1[df1['Start Time'].dt.month == 1]
df_month2 = df1[df1['Start Time'].dt.month == 2]
df_month3 = df1[df1['Start Time'].dt.month == 3]
df_month4 = df1[df1['Start Time'].dt.month == 4]
df_month5 = df1[df1['Start Time'].dt.month == 5]
df_month6 = df1[df1['Start Time'].dt.month == 6]
df_month7 = df2[df2['Start Time'].dt.month == 7]
df_month8 = df2[df2['Start Time'].dt.month == 8]
df_month9 = df2[df2['Start Time'].dt.month == 9]
df_month10 = df3[df3['Start Time'].dt.month == 10]
df_month11 = df3[df3['Start Time'].dt.month == 11]
df_month12 = df3[df3['Start Time'].dt.month == 12]
```

```python
outer_loop = True

while outer_loop:
    # Get city selection from user
    selected_city = input("Please select a city (New York City, Chicago or Washington): ")

    # Determine valid months based on current city
    valid_months = []
    if selected_city.lower() == 'new york city':
        valid_months = [1, 2, 3, 4, 5, 6]
    elif selected_city.lower() == 'chicago':
        valid_months = [7, 8, 9]
    elif selected_city.lower() == 'washington':
        valid_months = [10, 11, 12]
    else:
        print("Invalid city selection. Please choose New York City, Chicago or Washington.")
        continue

    inner_loop = True

    while inner_loop:
        # Select the relevant dataframe based on month selection
        selected_month = int(input(f"Please select a month between {', '.join(map(str, valid_months))} : "))
        if selected_month not in valid_months:
            print("Invalid month selection. Please select a valid month.")
            continue

        selected_df = globals()[f'df_month{selected_month}']

        # Show questions to user
        print("Questions:")
        print("1. Le jour de la semaine avec le plus d'activité")
        print("2. L'heure de démarrage la plus courante")
        print("3. La durée de voyage moyen sur la période (mois)")
        print("4. Le total pour chaque catégorie de User")
        print("5. Le nombre total de femmes et d'hommes sur la période")
        print("6. L'année de naissance la plus ancienne")
        print("7. L'année de naissance la plus récente")
        print("8. L'année de naissance la plus courante sur la période (avec le nombre d'occurence)")
        print("9. Exit (No)")

        # Kullanıcıdan soru seçimini al
        selected_question = input("Lütfen bir soru numarası seçin (1-9) veya çıkmak için 'No' yazın: ")

        # Take action according to the selected question
        if selected_question == '1':
            most_active_day = selected_df['Start Time'].dt.day_name().mode().values[0]
            print(f"Le jour de la semaine avec le plus d'activité est: {most_active_day}")
        elif selected_question == '2':
            most_common_hour = selected_df['Start Time'].dt.hour.mode().values[0]
            print(f"L'heure de démarrage la plus courante est: {most_common_hour} heures")
        elif selected_question == '3':
            total_trip_duration = selected_df['Trip Duration'].sum()
            total_records = len(selected_df)
            average_trip_duration = total_trip_duration / total_records
            print(f"La durée de voyage moyen sur la période (mois): {average_trip_duration} seconde")
        elif selected_question == '4':
            gender_counts = selected_df['Gender'].value_counts()
            print("Le total pour chaque catégorie de User:")
            print(gender_counts)
        elif selected_question == '5':
            gender_counts = selected_df['Gender'].value_counts()
            print("Le nombre total de femmes et d'hommes sur la période:")
            print(gender_counts)
        elif selected_question == '6':
            oldest_birth_year = selected_df['Birth Year'].min()
            print(f"L'année de naissance la plus ancienne est: {oldest_birth_year}")
        elif selected_question == '7':
            latest_birth_year = selected_df['Birth Year'].max()
            print(f"L'année de naissance la plus récente est: {latest_birth_year}")
        elif selected_question == '8':
            most_common_birth_year = selected_df['Birth Year'].mode().values[0]
            occurrence_count = selected_df[selected_df['Birth Year'] == most_common_birth_year].shape[0]
            print(f"L'année de naissance la plus courante sur la période est: {most_common_birth_year} avec {occurrence_count} occurrences")
        elif selected_question.lower() == 'no':
            print("The program is terminating. Have a nice day!")
            inner_loop = False
        else:
            print("Invalid question number. Please enter a number between 1 and 9 or 'No'.")

        # Ask the user if they want to continue
        continue_choice = input("Do you want to continue? (Yes/No): ")
        if continue_choice.lower() != 'yes':
            print("The program is terminating. Have a nice day!")
            outer_loop = False
            break
```
